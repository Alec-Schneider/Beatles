```python
import librosa
```

```python
song, sr = librosa.load("Because.mp4")
```

```python
mel = librosa.feature.melspectrogram(y=song, sr=sr)
```

```python
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1)
ax.set_xlabel("Time")
ax.set_ylabel("Hz")
librosa.display.specshow(librosa.power_to_db(mel, ref=np.max), y_axis="linear", x_axis="time")
plt.colorbar(format='%+2.0f dB')
```

```python
librosa.__version__
```

```python
stft = librosa.core.stft(song)
```

```python
librosa.display.specshow(librosa.power_to_db(stft, ref=np.max),  y_axis="linear", x_axis="time")
plt.colorbar(format='%+2.0f dB')
```

```python
song2, sr2 = librosa.load("Because.wav")
mel2 = librosa.feature.melspectrogram(y=song2, sr=sr2)
```
